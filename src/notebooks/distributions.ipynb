### All distributions

#### Discrete

- Bernoulli $(p)$
- Binomial $(n, p)$
- Poisson $(\lambda)$

#### Continuous

- Uniform $(a, b)$
- Normal (Gaussian) $(\mu, \sigma^2)$
- Exponential $(\lambda)$
- Lognormal $(\mu, \sigma^2)$
- Gamma $(k, \lambda)$
- Beta $(\alpha, \beta)$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

### Uniform Distribution

Models a value chosen equally likely from an interval $[a,b]$.

Notation: $X \sim U(a,b)$

PDF is constant: $f(x)=\dfrac{1}{b-a}$ on $[a,b]$.

CDF grows linearly from 0 to 1.

Mean: $\dfrac{a+b}{2}$ (center).

Variance: $\dfrac{(b-a)^2}{12}$ (depends on width).

```python
# Parameters
a = 0
b = 1

data = np.random.uniform(a, b, 100000000)

# Showing distribution
plt.figure(figsize=(12, 6))
plt.hist(data, bins=10, color='skyblue', alpha=0.7)
plt.title("Showing data distribution")
plt.xlabel("Bins")
plt.ylabel("data point")
plt.show()
```

### Normal/Gaussian Distribution

Models natural variation where values cluster around a mean.

Notation: $X \sim \mathcal{N}(\mu, \sigma^2)$

PDF:
$f(x)=\dfrac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\dfrac{(x-\mu)^2}{2\sigma^2}\right)$

CDF: integral of the PDF (no closed simple form).

Mean: $\mu$ (center).

Variance: $\sigma^2$ (spread).

```python
mu = 0
sigma = 2

data = np.random.normal(mu, sigma ** 2, 100000000)

plt.figure(figsize=(12, 6))
plt.hist(data, bins=100, color='skyblue', alpha=0.7)
plt.title("Showing data distribution")
plt.xlabel("Bins")
plt.ylabel("data point")
plt.show()
```

```python
_lambda = 10

data = np.random.exponential(_lambda, 100000)

plt.figure(figsize=(12, 6))
plt.hist(data, bins=100, color='skyblue', alpha=0.7)
plt.title("Showing data distribution")
plt.xlabel("Bins")
plt.ylabel("data point")
plt.show()
```

```python
mu = 0
sigma = 2

data = np.random.lognormal(mu, sigma, 100000)

plt.figure(figsize=(12, 6))
plt.hist(data, bins=10, color='skyblue', alpha=0.7)
plt.title("Showing data distribution")
plt.xlabel("Bins")
plt.ylabel("data point")
plt.show()
```
