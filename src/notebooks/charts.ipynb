### Importing all libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import random as rd
import numpy as np

sns.set_theme(style="whitegrid", context="talk", palette="bright")
```

### Bar Graphs

Generating random data for bar graph

```python
numbers = [rd.randint(0, 100) for _ in range(1000)]
categories = ['0-20', '21-40', '41-60', '61-80', '81-100']
frequency = [0, 0, 0, 0, 0]

def get_category_index(number):
    if 0 <= number <= 20:
        return 0
    elif 21 <= number <= 40:
        return 1
    elif 41 <= number <= 60:
        return 2
    elif 61 <= number <= 80:
        return 3
    elif 81 <= number <= 100:
        return 4

for number in numbers:
    idx = get_category_index(number)
    frequency[idx] += 1

# Create bar graph for frequency distribution
plt.figure(figsize=(12, 6))
plt.bar(categories, frequency, color='skyblue')
plt.title('Frequency Distribution of Random Numbers', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.ylim(0, max(frequency) + 50)
plt.show()
```

### Line Graphs

Plotting `sin(x)` and `cos(x)` functions

```python
# plotting from x = -2π to x = 2π

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
print(type(x))
```

```python
sine = np.sin(x)
cosine = np.cos(x)

print(type(sine))
print(type(cosine))

# Size
print(len(x))
print(len(sine))
print(len(cosine))
```

```python
# Plotting both sine and cosine functions
plt.figure(figsize=(14, 7))
plt.plot(x, sine, label='sin(x)', color='blue')
plt.plot(x, cosine, label='cos(x)', color='orange')
plt.title('Sine and Cosine Functions', fontsize=16)
plt.xlabel('x (radians)', fontsize=14)
plt.ylabel('Function Value', fontsize=14)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.legend()
plt.show()
```

### Plotting two subgraphs side by side

```python
# Two subplots for sine and cosine functions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 10))

# Plotting subplot for sine function
ax1.plot(x, sine, color='blue')
ax1.set_title('Sine Function', fontsize=16)
ax1.set_xlabel('x (radians)', fontsize=14)
ax1.set_ylabel('sin(x)', fontsize=14)
ax1.axhline(0, color='black',linewidth=0.5, ls='--')
ax1.axvline(0, color='black',linewidth=0.5, ls='--')
ax1.grid(True)

# Plotting subplot for cosine function
ax2.plot(x, cosine, color='orange')
ax2.set_title('Cosine Function', fontsize=16)
ax2.set_xlabel('x (radians)', fontsize=14)
ax2.set_ylabel('cos(x)', fontsize=14)
ax2.axhline(0, color='black',linewidth=0.5, ls='--')
ax2.axvline(0, color='black',linewidth=0.5, ls='--')
ax2.grid(True)  

plt.tight_layout()
plt.show()
```

### Pie Charts

```python
categories = ['A', 'B', 'C', 'D']
values = [120, 350, 210, 210]

# Create pie chart
plt.figure(figsize=(10, 10))
plt.pie(values, labels=categories, autopct='%.2f%%', startangle=140, colors=sns.color_palette("bright6"))
plt.title('Category Distribution', fontsize=16)
plt.show()
```

### Scatter Plots

```python
# scatter plot data
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100

# Create scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(x, y, color='green', alpha=0.9, edgecolors='w', s=100)
plt.title('Random Scatter Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.show()
```

### Histograms

```python
# Histogram data
data = np.random.rand(1000) * 100 # This array contains 1000 random numbers between 0 and 100

# Create histogram
plt.figure(figsize=(12, 8))
plt.hist(data, bins=4, color='purple', alpha=0.7)
plt.title('Histogram of Random Data', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

### Area Chart

```python
# Area Chart
x = np.linspace(0, 10, 100)
sine = np.sin(x) + 1  # Shift sine wave up to ensure all values are positive
cosine = np.cos(x) + 1  # Shift cosine wave up to ensure all values are positive

plt.figure(figsize=(12, 8))
plt.fill_between(x, sine, color='blue', alpha=0.5, label='sin(x) + 1')
plt.fill_between(x, cosine, color='orange', alpha=0.5, label='cos(x) + 1')
plt.title('Area Chart of Sine and Cosine Functions', fontsize=16)
plt.xlabel('x (radians)', fontsize=14)
plt.ylabel('Function Value', fontsize=14)
plt.legend()
plt.show()
```

### Heatmaps

This can be used if two variables are categorial and one is numerical.

```python
data = np.array([
    [120,  80,  40, 200],  # Mon
    [150,  90,  60, 220],  # Tue
    [180, 110,  70, 250],  # Wed
    [170, 130,  85, 270],  # Thu
    [220, 160, 120, 300]   # Fri
])

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
categories = ["Electronics", "Clothing", "Books", "Groceries"]

plt.figure(figsize=(10, 8))

sns.heatmap(
    data,
    annot=True,
    fmt="d",
    cmap="YlOrRd",
    xticklabels=categories,
    yticklabels=days
)

plt.title("Website Category Click Heatmap")
plt.xlabel("Product Category")
plt.ylabel("Day of Week")
plt.show()
```
